# Epic 3: Data Pre-Processing

## Objective

Data preprocessing prepares the dataset for machine learning by handling missing values, removing inconsistencies, encoding categorical variables, balancing the dataset, scaling numerical features, and splitting the data into training and testing sets.

## Story 1: Data Cleaning

The dataset is checked for missing values, duplicate records, and inconsistencies. Missing values are handled appropriately to ensure data quality before model training.

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')

plt.title("Missing Values Heatmap")

plt.show()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

df['Married'] = df['Married'].fillna(df['Married'].mode()[0])

df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])

df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])

df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])

df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])

df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)

print("Remaining Rows :", df.shape[0])
```

### Observation

- Missing values were successfully handled.
- Numerical missing values were replaced using the median.
- Categorical missing values were replaced using the mode.
- Duplicate records were removed.
- The dataset is now clean and ready for preprocessing.

## Story 2: Dataset Balancing

Before balancing, the target variable distribution is examined to determine whether the dataset is imbalanced.

```python
df['Loan_Status'].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Loan_Status', data=df)

plt.title("Loan Status Distribution")

plt.show()
```

```python
df['Loan_Status'].value_counts(normalize=True)*100
```

### Observation

The target classes are reasonably balanced. Therefore, no oversampling or undersampling technique was applied, and the original class distribution was retained for model training.

## Story 3: Encoding and Feature Scaling

Machine learning algorithms require numerical input. Therefore, categorical variables are encoded into numerical values, and numerical features are scaled to improve model performance.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_columns = [
    'Gender',
    'Married',
    'Dependents',
    'Education',
    'Self_Employed',
    'Property_Area',
    'Loan_Status'
]

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

df.head()
```

```python
df.drop('Loan_ID', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_columns = [
    'ApplicantIncome',
    'CoapplicantIncome',
    'LoanAmount',
    'Loan_Amount_Term'
]

df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

df.head()
```

```python
import pickle

pickle.dump(scaler, open("scaler.pkl", "wb"))
```

### Observation

Categorical variables were encoded using Label Encoding. Numerical variables were standardized using StandardScaler to improve the performance of machine learning algorithms.

## Story 4: Splitting the Dataset

The processed dataset is divided into training and testing sets to evaluate the performance of machine learning models on unseen data.

```python
X = df.drop('Loan_Status', axis=1)

y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print("Training Labels :", y_train.shape)
print("Testing Labels  :", y_test.shape)
```

# Epic 3 Summary

The dataset was successfully preprocessed by handling missing values, removing duplicates, encoding categorical variables, scaling numerical features, and splitting the dataset into training and testing sets. These preprocessing steps ensure that the data is clean, consistent, and suitable for training machine learning models.

# Epic 4: Model Building

## Story 1: Decision Tree Classifier

A Decision Tree classifier is trained using the preprocessed dataset to predict loan approval status. The model is evaluated using classification metrics.

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)
```

```python
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)
```

```python
dt_pred = dt_model.predict(X_test)
```

```python
dt_accuracy = accuracy_score(y_test, dt_pred)

print("Decision Tree Accuracy :", round(dt_accuracy*100,2),"%")
```

```python
cm = confusion_matrix(y_test, dt_pred)

plt.figure(figsize=(5,4))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title("Decision Tree Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
print(classification_report(y_test, dt_pred))
```

### Observation

The Decision Tree model successfully classified loan applications with good accuracy. However, it may overfit the training data due to its tree-based structure.

## Story 2: Random Forest Classifier

Random Forest combines multiple decision trees to improve prediction accuracy and reduce overfitting.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
rf_pred = rf_model.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, rf_pred)

print("Random Forest Accuracy :", round(rf_accuracy*100,2),"%")
```

```python
cm = confusion_matrix(y_test, rf_pred)

plt.figure(figsize=(5,4))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Greens'
)

plt.title("Random Forest Confusion Matrix")

plt.show()
```

```python
print(classification_report(y_test, rf_pred))
```

### Observation

Random Forest generally performs better than a single Decision Tree by reducing variance and improving prediction stability.

## Story 3: K-Nearest Neighbors (KNN)

The K-Nearest Neighbors (KNN) algorithm classifies a loan application based on the similarity of applicant features with previously observed data.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train, y_train)
```

```python
knn_pred = knn_model.predict(X_test)
```

```python
knn_accuracy = accuracy_score(y_test, knn_pred)

print("KNN Accuracy :", round(knn_accuracy*100,2), "%")
```

```python
cm = confusion_matrix(y_test, knn_pred)

plt.figure(figsize=(5,4))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Oranges'
)

plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
print(classification_report(y_test, knn_pred))
```

### Observation

The KNN model classifies applicants based on feature similarity. Its performance depends on the selected value of K and the scaling of numerical features.

## Story 5: Model Evaluation and Best Model Selection

All trained models are compared based on their training and testing accuracy. The best-performing model is saved for deployment in the Flask application.

```python
dt_train = dt_model.score(X_train, y_train)

rf_train = rf_model.score(X_train, y_train)

knn_train = knn_model.score(X_train, y_train)

xgb_train = xgb_model.score(X_train, y_train)
```

```python
dt_test = dt_model.score(X_test, y_test)

rf_test = rf_model.score(X_test, y_test)

knn_test = knn_model.score(X_test, y_test)

xgb_test = xgb_model.score(X_test, y_test)
```

```python
comparison = pd.DataFrame({

    "Model":[
        "Decision Tree",
        "Random Forest",
        "KNN",
        "XGBoost"
    ],

    "Training Accuracy":[
        dt_train,
        rf_train,
        knn_train,
        xgb_train
    ],

    "Testing Accuracy":[
        dt_test,
        rf_test,
        knn_test,
        xgb_test
    ]

})

comparison
```

```python
comparison["Training Accuracy"] = comparison["Training Accuracy"]*100

comparison["Testing Accuracy"] = comparison["Testing Accuracy"]*100

comparison.round(2)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Model",
    y="Testing Accuracy",
    data=comparison
)

plt.title("Model Comparison")

plt.ylabel("Accuracy (%)")

plt.show()
```
